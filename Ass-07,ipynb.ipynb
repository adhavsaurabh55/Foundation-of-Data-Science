```python
# Importing required libraries
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y, color='blue', marker='o', linestyle='--', label='Line Plot')

# Adding title and labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show legend
plt.grid(True)  # Add grid
plt.show()
```

```python
# Data
categories = ['Math', 'Physics', 'Chemistry', 'Biology']
marks = [80, 75, 90, 85]

# Create bar plot
plt.bar(categories, marks, color='green')

# Add title and labels
plt.title("Marks of Students in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# Sample data
ages = [18, 19, 20, 18, 21, 22, 20, 19, 21, 22, 23, 21]

# Create histogram
plt.hist(ages, bins=5, color='orange', edgecolor='black')

# Add title and labels
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Sample data
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

# Create scatter plot
plt.scatter(x, y, color='red', marker='x')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Seaborn library

```python
# Import seaborn
import seaborn as sns
import pandas as pd

# Sample dataset
data = {'Student': ['A', 'B', 'C', 'D'],
        'Marks': [80, 75, 90, 85]}

df = pd.DataFrame(data)

# Bar plot using seaborn
sns.barplot(x='Student', y='Marks', data=df, palette='Blues_d')
plt.title("Seaborn Bar Plot")
plt.show()
```

```python
# Sample data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temp = [22, 24, 21, 23, 25]

sns.lineplot(x=days, y=temp, marker='o', color='purple')
plt.title("Temperature Over a Week")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
ages = [18, 19, 20, 18, 21, 22, 20, 19, 21, 22, 23, 21]

# Create histogram
sns.histplot(ages, bins=5, color='skyblue', kde=True)  # kde=True adds smooth curve

# Add title and labels
plt.title("Age Distribution (Seaborn Histogram)")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Sample data
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

# Create scatter plot
sns.scatterplot(x=x, y=y, color='red', s=100)  # s=100 sets size of points

# Add title and labels
plt.title("Scatter Plot Example (Seaborn)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (correlation-like table)
data = {
    'Math': [80, 75, 90, 85],
    'Physics': [70, 78, 88, 90],
    'Chemistry': [85, 80, 92, 87],
    'Biology': [75, 85, 80, 88]
}

# Convert to DataFrame
df = pd.DataFrame(data, index=['Student A', 'Student B', 'Student C', 'Student D'])

# Create heatmap
sns.heatmap(df, annot=True, cmap='YlGnBu', linewidths=0.5)

# Add title
plt.title("Students Marks Heatmap")
plt.show()
```
